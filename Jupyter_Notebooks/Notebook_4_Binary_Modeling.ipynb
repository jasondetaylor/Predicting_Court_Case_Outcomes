### One Hot Encoding (OHE)

Nest we need to convert our categorical variables into a language that can be fed into our models: numbers. We will use OHE to do this. Briefly, OHE converts each categorical columns into a binary vector where all the elements are zero except for the index corresponding to the category, which is set to one. Thus our categorical columns of strings become binary columns that can be interpreted in out ML models.

First reset the indexes of our split data:

```python
X_test = X_test.reset_index(drop = True)
X_train = X_train.reset_index(drop = True)
y_train = y_train.reset_index(drop = True)
```

Then create new dataframes containg only our categorical data to be transformed:

```python
# define a dataframe containing our categorical data:
cat_df_train = X_train.select_dtypes(include = 'object')

# note we must reset the index of our test data to begin row index at zero to align with our encoded data
cat_df_test = X_test.select_dtypes(include = 'object')
```

```python
cat_df_train.head()
```

```python
cat_df_test.head()
```

Next perform our OHE. Note this is done separately on our train and test data to avoid any data leakage.

```python
# Instantiate
ohe_train = OneHotEncoder()
ohe_test = OneHotEncoder()

# fit and transform
encoded_train = ohe_train.fit_transform(cat_df_train)
encoded_test = ohe_test.fit_transform(cat_df_test)

# Convert from sparse matrix to dense
dense_array_train = encoded_train.toarray()
dense_array_test = encoded_test.toarray()

# Put into a dataframe to get column names
encoded_train_df = pd.DataFrame(dense_array_train, columns = ohe_train.get_feature_names_out(cat_df_train.columns), dtype='int16').reset_index(drop=True)
encoded_test_df = pd.DataFrame(dense_array_test, columns = ohe_test.get_feature_names_out(cat_df_test.columns), dtype='int16')
```

Note we are not dropping a col here, instead we use VIF to weed out multicollinearities in a smarter way. It also leaves all the variables in our base 'X_train' and 'X_test' for use in models where multicollinearity is not a concern.

```python
encoded_train_df.head()
```

```python
encoded_test_df.head()
```

We can now combine our newly created OHE dataframes with our numeric variables to reform our X dataframes:

```python
## redefine our independent variable sets using this new cenoded data combined with the numeric data
X_train = pd.concat([encoded_train_df, X_train.select_dtypes(include = 'number')], axis = 1).reset_index(drop = True)
X_test = pd.concat([encoded_test_df, X_test.select_dtypes(include = 'number')], axis = 1).reset_index(drop = True)
```

### Multicollinearity

Lastly we need to remove features that cause multicolinearity issues within our dataframe. Firstly lets plot out a correlation heatmap:

```python
plt.figure(figsize=(20,10))
matrix = np.triu(X_train_scaled.corr())
sns.heatmap(X_train_scaled.corr(), cmap = 'coolwarm', mask = matrix, vmin = -1, vmax = 1, annot = True)
plt.show()
```

As is often the case with larger dataframes this plot is unclear, however we can see there are some issues between some variabes that require further investigating. Before we proceed, lets make a copy of our dataframes to retain a set that has all variables present:

```python
# copy variables over, note 'nomc' stands for no multicollinearity
X_train_scaled_nomc = X_train_scaled.copy()
X_test_scaled_nomc = X_test_scaled.copy()
```

We can analyze our correlation matrix using a VIF test:

```python
## asumming same patterns that are present in train are also in test
VIF_test = pd.Series([variance_inflation_factor(sm.add_constant(X_train_scaled_nomc).values, i) 
               for i in range(sm.add_constant(X_train_scaled_nomc).shape[1])], 
               index=sm.add_constant(X_train_scaled_nomc).columns)[1:] # leaving out the constant

VIF_test.sort_values(ascending = False).head()
```

Generally, VIF score greater than 5 are considered high. We can see we have some highly problematic features within our dataframe. This is to be expected as we did not drop any columns after OHE our categorical data, hence the variable shown above are a result of this. We can now choose which variables to drop based on their interference with the rest of the dataframe as calculated by the VIF test. Let's setup a while loop to redact the variable with the highest VIF score and then rerun the test until all VIF scores are less than 5. Note we are making the assumption here that variables found to have a high VIF score on the train set can also be dropped from the test set.

```python
while VIF_test.max() > 5:
    #drop variable with highest VIF score
    X_train_scaled_nomc.drop(VIF_test.sort_values(ascending = False).index[0], axis = 1, inplace = True)
    X_test_scaled_nomc.drop(VIF_test.sort_values(ascending = False).index[0], axis = 1, inplace = True)
    
    #rerun VIF test
    VIF_test = pd.Series([variance_inflation_factor(sm.add_constant(X_train_scaled_nomc).values, i) 
               for i in range(sm.add_constant(X_train_scaled_nomc).shape[1])], 
               index=sm.add_constant(X_train_scaled_nomc).columns)[1:]
    
VIF_test.sort_values(ascending = False).head()
```

We can see that now our highest VIF score is now 2.05. Our redated dataframe now has a shape:

## Modeling

### Pipelines and GridSearch

As mentioned in section 3, we were unable to utilize a pipe to it's full potential due to a package issue. We will still make use of pipelines here but only in conjuction with GridSearch for hyperparameter tuning and not data preprocessing. The benefit of using pipelines here is that they will automatically apply a k-fold within the GridSearch to allow us to easily optimize our hyperparameters. 

The following section makes use of dictionaries to run a separate pipeline for each of:
- Logistic Regression
- Desicion Trees
- K Nearest Neighbors

Note that and Support Vector Machine (SVM) was also planned to be run however the model diverges for unknown reasons and has been redacted from this study. The models are evaluated with respect to over or under fitting to the train data. A brief description of the mehtodology of each of these methods is given below:

Logistic Regression: <br>
The base of a logistic regression is a linear regression, thus we must first understand a linear regression. A linear regression model works to find a linear line of best fit by applying a coefficient to each independent varaible. The line that represents the best-fit relationship between the independent variables and the target variable is found using a technique called least squares regression. The algorithm finds the values of the coefficients that minimize the sum of the squared differences between the predicted values and the actual values of the target variable. Logistic regression works by passing in the results of a linear regression into a sigmoid function to map the result to a probability between zero and one. This output is known as a 'soft prediction'. A threshold is then applied to the 'soft prediction' to convert each probability into either a 1 or a zero, thus producing a 'hard prediction'. Note that this is the case for binary problems, however as an aside multiclass classification problems operate on a 'one vs rest' basis where the prediction for each class is treated as a binary problem where the class being predicted is the target class (class 1) and all other classes are class 0. Thus we end up with a binary classification for each unidque target class as our 'soft prediction'. The concept of a threshold does not directly apply to multiclass problems as we take the maximum soft prediction value to identify the most probably class for each instance which gives us our hard prediction.

SVM: <br>
An SVM is, similar to logistic regression, a linear model with coefficients that acts to create a boundary between classes to form predictions. However SVM's fall under the class of 'maximum margin classifiers' which is where it differs from logistic regression. An SVM model utilizes the coefficients to create a boundary that maximises the difference between this boundary and the data. This can help to deal with degrees of variance in the test data. SVM's can also be applied as non linear classifiers through what is known as the 'kernel trick' where the dataset is converted to a higher dimension. The SVM still applies a linear boundary to the data however this gives non-linear boundary when transformed back to the original space. This can be useful when trying to resolve patterns that are not linearly seperable as is often the case.

Decision Trees: <br>
Decision trees are non-linear, non-distance base models which learn by applying multiple binary classifications in a 'tree-like' methodology made up of nodes. The 'root' node is the topmost node in the tree, representing the starting point of the decision-making process. The root node is connected to 'internal' nodes via edges or branches. Each internal node represents a linear decison boundary of the model that splits the data using the variable and value that has the most value, i.e best resolves th epatterns in the data at that point. This process is continued forming multiple nodes and branches until we arrive at a 'leaf' node containing are the final predictions of the model. 

K Nearest Neigbors: <br>
KNN is another non-linear model, but calculates distances between data points to come with classifications. Unlike our other models, KNN does not train a model and utilize it to make preictions. Instead, each new data point is assessed with respect to the 'cloud' of training data. KNN then calculates the distance between the new data point and every point in our training data. Based on the the label of the points nearest to it. The number of nearest points considered is set by the 'n_neighbors' hyperparameter giving it it's name 'K Nearest Neighbors'. Note this model can be computationally heavy due to the large number of distances that need to be caluclated.

First define the range of hyperparameters we want to optimize over, making use of a round function:

```python
def round_to_nearest_100(num):
    return round(num / 100.0) * 100
```

```python
# C values for logreg
C_values = np.logspace(-6, 6, 13)

# n_neighbors for KNN
n_max = round_to_nearest_100(np.sqrt(X_train_scaled.shape[0]))
n_step = 50
n_neighbors = np.arange(n_step, n_max, n_step)

# max_depths for DTree
d_max = round_to_nearest_100(np.sqrt(X_train_scaled.shape[0]))
d_step = 20
max_depths = np.arange(d_step, d_max, d_step)
```

Now set up our dictionaries allowing up to iterate to run each pipeline for each model with the corresponding dataframe and hyperparameters. Each pipe/model is hen passed into a grid search to optimize our hyperparameters.

```python
# note SVM diverging, excluded from study

# Define dataframes
X_dict = {'logreg': X_train_scaled_nomc,
          #'SVM': X_train_scaled_nomc,
          'KNN': X_train_scaled,
          'DTree': X_train
         }

# separate pipelines for each model
pipelines = {'logreg': Pipeline([('logreg', LogisticRegression())]),
             #'SVM': Pipeline([('SVM', svm.SVC())])
             'KNN': Pipeline([('KNN', KNeighborsClassifier())]),
             'DTree': Pipeline([('DTree', DecisionTreeClassifier())])
            }

# Set up hyperparameters for all models
params = {'logreg': {'logreg__C': C_values},
          #'SVM': {'SVM__C': [0.1, 1, 10], 'SVM__kernel': ['linear', 'rbf']},
          'KNN': {'KNN__n_neighbors': n_neighbors},
          'DTree': {'DTree__max_depth': max_depths}
          }


# Set up grid search for each pipeline
results_dict = {}
for key, X in X_dict.items():
    pipe = pipelines[key]
    gs = GridSearchCV(pipe, params[key])
    results_dict[key] = gs.fit(X, y_train)
```

#### Model Evaluation

Evaluate our output of our gridsearch by scoring our model using their chosen hyperparameters. Similar to above we define a dictionary of dataframes for our X_test to align the dataframe to the corresponding model:

```python
# Define dataframes
X_test_dict = {'logreg': X_test_scaled_nomc,
               'KNN': X_test_scaled,
               'DTree': X_test}
```

Iterate through our grid search results to show the hyperparamter and score our test and train data:

```python
for key, results in results_dict.items():
    print(f"Results for {key}:")
    print(f"Best parameters: {results.best_params_}")
    print(f"Train accuracy: {results.score(X_dict[key], y_train):4f}")
    print(f"Test accuracy: {results.score(X_test_dict[key], y_test):4f}\n")
```

The results for Logistic Regression and K Nearest Neighbors is good, with similar train and test scores suggesting our models have been appropriately fitted to our train data resulting in no under or over fitting. Our Decision Tree however is seen to significantly overfit to the train data suggesting. An investigation into this is covered in section 5.1.2 following.

#### Decision Tree Investigation

Decision trees overfit when the max_depth is set too high, allowing the model to conform to intricate patterns specific to the training data, leading to poor performance on test data. Upon further inspection to value chosen by our gridsearch for the max_depth was the lowest value passed in and is still overfitting. This surprising as a general rule of thumb we usually aim for a max_depth value equal to the sqaure root of the number of rows in our dataframe, as was passed into the gridsearch.

Lets do a manual investigation to check our diagnosis here by running a fine hyperparamter tune up to max_depth of 50:

```python
#Store the results
cross_validation_scores = []
max_depths = np.arange(2, 51, 2)
#Do some cross validation
for max_depth in max_depths:
    
    # instantiate
    DTree = DecisionTreeClassifier(max_depth = max_depth, random_state = 42)
    
    # the cross validation score (mean of scores from all folds)
    cv_score = np.mean(cross_val_score(DTree, X_train, y_train, cv = 5))
    
    # append each score to a list
    cross_validation_scores.append(cv_score)
```

One note here is the inclusion of a random state which was not done in the pipeline, unsure if there is a way to set a random state for pertinent models in a pipeline.

```python
optimal_depth = max_depths[np.argmax(cross_validation_scores)]

sns.lineplot(x = max_depths, y = cross_validation_scores, label = 'Average Cross Validation Score')
plt.axvline(x = optimal_depth, color = 'black', linestyle = '--', label = f'Optimal Depth = {optimal_depth}')
plt.legend()
plt.show()
```

This would suggest that max depth only needs 10 features to optimally distingyish between our 2 classes, after that it appears we start to fit to patterns only present in our trainign data. This number is much lower than expected as mentioned above as the rule of thumb for the max_dpeth hyperparamter is the sqrt of the number of rows as was initially inputted into our pipe. ### Advanced Methods

We will now apply some more advanced methods to see if they can further resolve our data, namely will will work with:
- Extreme Gradient Boosting (XG Boost)
- Neural Network

A breif summary of the moethodologies of these methods is given below.

XG Boost: <br>
XG Boost is an ensemble method as it combines the predictions of multiple models. It works by iteratively building a series of decision trees, with each new tree trained to correct the errors of the previous trees. Am important note is that this differs from Random Forest models as Random Forests are a large number of decision trees run independently with the prediction being average of majority vote of all the trees. In contrast, XGBoost builds decision trees sequentially, with each new tree trained to correct the errors of the previous trees.

Neaural Networks: <br>
Neural networks, on the other hand, are a type of machine learning algorithm that are designed to model complex nonlinear relationships. They consist of interconnected layers of nodes (neurons) that process input data and produce output predictions. Each neuron computes a weighted sum of its inputs, applies a non-linear activation function to the sum, and passes the result to the next layer of the network. Neural networks are trained using a technique called backpropagation, which adjusts the weights of the network to minimize a loss function that measures the difference between the predicted and actual outputs.

```python
def VIF_test(X):
    VIF_test = pd.Series([variance_inflation_factor(sm.add_constant(X).values, i) 
               for i in range(sm.add_constant(X).shape[1])], 
               index=sm.add_constant(X).columns)[1:]
    return VIF_test.sort_values(ascending = False)
```

```python
def VIF_feature_selection(X, threshold):
    '''
    Takes a dataframe and a threshold value.
    Returns a new dataframe 'X_nomc' with variables removed with a VIF score higher that the threshold.
    Note if all VIF scores of input dataframe are less than the threshold, function will return print
    statement only and leave the dataframe unchanged.
    '''
    
    # first run VIF test on df:
    VIF_scores = VIF_test(X)
    
    if VIF_scores.max() <= threshold:
        return 'All VIF scores less than threshold, dataframe unchanged'
    else:
        # copy df over to leave original unchanged
        X_nomc = X.copy()
        
        while VIF_scores.max() > threshold:
            #drop variable with highest VIF score
            X_nomc.drop(VIF_test.sort_values(ascending = False).index[0], axis = 1, inplace = True)

            #rerun VIF test
            VIF_scores = VIF_test(X_nomc)
        return X_nomc
```
