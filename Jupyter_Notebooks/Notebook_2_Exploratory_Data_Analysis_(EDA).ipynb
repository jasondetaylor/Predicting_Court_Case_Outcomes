## Introduction

Performing Exploratory Data Analysis (EDA) is an important step in our project. Chiefly, EDA allows us to gain a better understanding of our data. This includes identifying patterns, trends, and outliers in the data, as well as getting a sense of the data distribution, its range, and its general characteristics. Fully understanding the data is key in effectively channeling the modelling efforts to follow. EDA is also where we have engineered some additional features. In our case, this involved the creation of our target varaible as well as other variables that may help us resolve and identfy patterns in our data. This notebook illustrates the EDA performed on our cleaned dataframe recieved from our previous notebook specifically:
- Feauture Engineering
- Data Distributions
- Target Variable Relationships
- Time Variable Analysis

The output of this notebook is a new dataframe containing additonal features which is passed to both our modeling notebooks (notebooks 3 and 4).

## Feature Engineering

### Target Variable

```python
#create new answer key (note second np.where is the else of the first np.where)
#key: 0 = sentence is below guideline, 1 = sentence is within guideline, 2 = sentence is above guideline
df3['sentence_guideline_diff'] = np.where(
    (df3['SENTTOT'] <= df3['GLMAX']) & (df3['SENTTOT'] >= df3['GLMIN']), 1, 
    np.where(df3['SENTTOT'] > df3['GLMAX'], 2, 0)
)

df3['sentence_guideline_diff'] = df3['sentence_guideline_diff'].astype('int8')
df3.drop(['SENTTOT', 'GLMIN', 'GLMAX'], axis = 1, inplace = True)
```

```python
df3['sentence_guideline_diff'].unique()
```

### Time variables

Create numeric columns for date variables from our 'DISPDATE' column:

```python
# year
df3['year'] = df3['DISPDATE'].dt.year
# month
df3['month'] = df3['DISPDATE'].dt.month
# day of month
df3['day_of_month'] = df3['DISPDATE'].dt.day
# day of week
df3['day_of_week'] = df3['DISPDATE'].dt.weekday
```

```python
# sanity check
df3.head()
```

### Age of Judge

USe our year column to calculate the age of the judge at the time of sentencing:

```python
df3['age_of_judge'] = df3['year'] - df3['BirthYear']
df3.drop('BirthYear', axis = 1, inplace = True)
# sanity check
df3['age_of_judge']
```

### Time Variable Analysis

Let's dig further into our time variables to better understand how our other variable may have morphed over time and see if there are any cyclical patterns we can idenify.

```python
# define our time variables:
time_variables = ['year', 'month', 'day_of_month', 'day_of_week']
```

#### Target Variable

```python
ncols = 2
nrows = int(len(time_variables) / ncols)
plt.subplots(ncols = ncols, nrows = nrows, figsize = (10, 7))
colours = sns.color_palette('Set1', n_colors=len(df3['sentence_guideline_diff'].unique()))

for i, timespan in enumerate(time_variables, 1):
        
        data_df = df3.groupby([timespan, 'sentence_guideline_diff']).size().reset_index(name='count')

        sums = data_df.groupby([timespan, 'sentence_guideline_diff'])['count'].sum()

        # calculate the total sum for each day of the week
        total_sums = sums.groupby('sentence_guideline_diff').sum()

        # calculate the percentage for each count
        data_df['percentage'] = data_df.apply(lambda row: row['count'] / total_sums[row['sentence_guideline_diff']] * 100, axis=1)
        
        plt.subplot(ncols, nrows, i)
        sns.lineplot(data = data_df, x = timespan, y = 'percentage', hue = 'sentence_guideline_diff', palette = colours)
        plt.title(f'Target Variable Variation by {timespan}')
        plt.ylabel('Percentage')
        plt.xlabel(timespan)
plt.tight_layout()
```

This plot shows how each of our target classes is distributed across each of our time spans. Note these plots can be interpreted as out of all my class 'X' rows, what proportion of them occur on each time value (i.e day of th week/month etc). The sum off all the points for each curve all up to 100%. This lacks some interpretability as the plots are dependent on the number of cases in that timespan. Lets look at the differecne between the lines at each timespan:

```python
ncols = 2
nrows = int(len(time_variables) / ncols)
plt.subplots(ncols = ncols, nrows = nrows, figsize = (10, 7))
colours = sns.color_palette('Set1', n_colors=len(df3['sentence_guideline_diff'].unique()))

for i, timespan in enumerate(time_variables, 1):
        
        data_df = df3.groupby([timespan, 'sentence_guideline_diff']).size().reset_index(name='count')
        
        # calculate the total rows for each time span
        total_rows = df3.groupby(timespan).size()

        # normalize count column by the corresponding value for each time span
        data_df['count_normalized'] = data_df.apply(lambda row: row['count'] / total_rows[row[timespan]], axis=1)

        sums = data_df.groupby([timespan, 'sentence_guideline_diff'])['count'].sum()

        # calculate the total sum for each sentence guideline difference
        total_sums = sums.groupby('sentence_guideline_diff').sum()

        # calculate the percentage for each count
        data_df['percentage'] = data_df.apply(lambda row: row['count_normalized'] / total_sums[row['sentence_guideline_diff']] * 100, axis=1)
        
        plt.subplot(ncols, nrows, i)
        sns.lineplot(data = data_df, x = timespan, y = 'percentage', hue = 'sentence_guideline_diff', palette = colours)
        plt.title(f'Target Variable Variation by {timespan}')
        plt.xlabel('Average Sentence Guideline Difference')
        plt.ylabel(timespan)
plt.tight_layout()
```

This gives a normalized view of the patterns of our target variable across different timespans. We can see that overall our class 2 shows more sporadic patterns due to the low number of samples in this class. Our year plot shows our stringest trend (by scale) and shows the number of instances of class 0 and class 2 decisions (lenient and harsh respectively) have been increasing relatively linearly over the years, with class 1 decisions (standard) dropping linearly to match. This suggests that over time, judges are becoming less reliant on and/or are more willing to venture outside of the recommended guidelines.

Additionally,  we can see that the scale of the month, day_of_month and day_of_week plots are all relatively small compared to the year. In spite of this, we can see some interesting patterns in month and day_of_week where the end of the month and the middle of the week tend to have a higher proportion of class 2 cases (harsh sentences).

#### Numeric Variables

```python
# define numeric variables for time study
numeric_variables = [col for col in numeric_cols if col not in time_variables]
```

```python
ncols = 2
nrows = int(len(time_variables) / ncols)
plt.subplots(ncols = ncols, nrows = nrows, figsize = (10, 7))
for i, timespan in enumerate(time_variables):
    for col in numeric_variables:
        grouped = df3[col].groupby(df3[timespan]).mean()
        data = grouped / grouped.max()
        plt.subplot(ncols, nrows, i+1)
        sns.lineplot(x = data.index, y = data.values, label = col)
    plt.title(timespan)
    plt.legend()
plt.tight_layout()
```

Again, year is a strong variable showing a reltively strong increase in all but out NUMDEPEN variable over time. As before our month and day_of_week timespans show very weak patterns as indicated by the very small scale. Another note is the sharp increase of out INT2 variable towards the end of the month. INT2 is the number of days between the court proceedings and the disposition date. The reason for this pattern in unknown.

#### Categorical Variables

```python
ncols = 2
nrows = int((len(categorical_cols) * len(time_variables)) / ncols)

plt.figure(figsize=(10, 50))
plt.subplots_adjust(hspace=0.5)

for i, col in enumerate(categorical_cols):
    for j, timespan in enumerate(time_variables):
        plt.subplot(nrows, ncols, i * len(time_variables) + j + 1)
        grouped_df = df3[col].groupby(df3[timespan]).value_counts().rename('count').reset_index()
        grouped_df['count'] = grouped_df['count'] / grouped_df['count'].max()
        sns.lineplot(data = grouped_df, x = timespan, y = 'count', hue = col)
        plt.title(f'{col} vs {timespan}')
        
        # in the interest of sapce, only show legend for every second plot, on right hand side (applies to both plots in row)
        if j % 2 == 0:
            plt.legend().remove()
        else:
            plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

The relative sizes of each of the categories make these plots difficult to inerpret and may not be ideal for this kind of analysis. No findings here.
